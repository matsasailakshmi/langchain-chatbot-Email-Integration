```python
import os
import smtplib
from email.message import EmailMessage
from getpass import getpass
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
```

```python
os.environ["OPENAI_API_KEY"] = getpass("Enter your OpenRouter API Key: ")
```

```python
llm = ChatOpenAI(
    model="openai/gpt-4o",
    base_url="https://openrouter.ai/api/v1",
    max_tokens=500,
    temperature=0.7
)
```

```python
def send_email(receiver_email, message_content):
    sender_email = input("Enter your Gmail address: ")
    app_password = getpass("Enter your Gmail App Password: ")

    msg = EmailMessage()
    msg["From"] = sender_email
    msg["To"] = receiver_email
    msg["Subject"] = "Chatbot Response"
    msg.set_content(message_content)

    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
        server.login(sender_email, app_password)
        server.send_message(msg)

    print("✅ Email sent successfully!")
```

```python
print("🤖 Chatbot Ready! Type 'exit' to stop.\n")

while True:
    question = input("You: ")

    if question.lower() == "exit":
        print("Goodbye 👋")
        break

    receiver_email = input("Enter receiver email: ")

    # Get LLM response
    response = llm.invoke([HumanMessage(content=question)])

    print("Chatbot:", response.content)

    # Send response via email
    send_email(receiver_email, response.content)
```
